# Functia scatter pune punctele imprastiate

```python
lista_1 = [1,2,3]
lista_2 = [4,5,6]

plt.scatter(lista_1,lista_2);
plt.show()
```

```python
lista_1 = [1]
lista_2 = [4]

plt.scatter(lista_1,lista_2);
plt.show()
```

```python
range_1 = range(4,8)
range_2 = range(6,10)

plt.scatter(range_1,range_2);
plt.show()
```

```python
range_1 = range(4)
range_2 = [3] * 4
plt.scatter(range_1,range_2);
plt.show()
```

# plt.title = titlu

```python
x = [10,10]
y = [8, 9]
plt.tilte('Acesta este titlul')
plt.scatter(x,y);
plt.show()
```

# X, Y 

```python
x = [10,10]
y = [8, 9]
plt.title('Acesta este titlul')
plt.xlabel("Valoarea X")
plt.ylabel("Valoarea Y")
plt.scatter(x,y);
plt.show()
```

```python
a = [10,10, 11, 20]
b = [8, 9, 10, 10]
plt.title('Acesta este titlul')
plt.xlabel("Valoarea X")
plt.ylabel("Valoarea Y")
plt.scatter(a,b);

plt.savefig("Figura mea")
```
